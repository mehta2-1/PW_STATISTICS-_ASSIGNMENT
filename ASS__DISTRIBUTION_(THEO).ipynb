#Q4.What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.

#ANS-

## STANDARDIZATION is the process of converting raw data into Z-SCORE , Whic tells how many  stand.D a value is away from the mean.
##FORMULA : Z= (ORIGINAL VALUE - MEAN) /STANDARD DEVIATION.
##IMPORTANCE:1.Allows comparison of values from different datasets
###           2.Removes the effect of units  of measurement.                         
###            3.Helps identify outliers.
###           4.Required for many statistical methods (e.g., normal probability calculations, machine learning models).
## TRANSFORMATION OF DATA : 1.Original mean = 0.
###                         2.Original standard deviation = 1.
###                         3.Shape of distribution remains the same.
## EXAMPLE : a student scores 70 in a test
            Mean = 60
            Standard deviation = 5
            =(70 - 60)/5
            =2 ,This means the score is 2 standard deviations above the mean.
